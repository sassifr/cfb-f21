This is an example of the type of data visualization that will allow us to visualize the relationships between multiple variables.

```python
import pandas as pd
import seaborn as sns
```

Seaborn is just a nicer plotting library, since matplotlib and the traditional MATLAB plots are quite obtuse and unappealing (IMO).

```python
# load the data
game_data = pd.read_csv("../data/game_stats.csv")
games = pd.read_csv("../data/games.csv")
games
```

```python
game_data
```

I will take a look at the relationship between the total number of yards and point differential. First, I will merge the two tables.

```python
combined_data = games.merge(game_data, left_on="id", right_on="id", validate="one_to_one")
combined_data
```

Now we will generate the Point differential. Here, we subtract the scores scored by the away team from the home team.

```python
def point_diff(row):
    return row["home_points"] - row["away_points"]


combined_data["point_differential"] = combined_data.apply(point_diff, axis=1)
combined_data["point_differential"]
```

Now that we have point differential, we will plot it against the total yards for both home and away teams.

```python
sns.scatterplot(
    data=combined_data, x="home.totalYards", y="point_differential"
).set_title("Home Team Total Yards vs Point Differential")
```

From this plot, we can see that the more total yards the home team has, the higher the point differential. This makes sense since the more total yards a team has, the more likely that they moved the ball down the field towards the endzone resulting in points. We should expect the inverse relationship for the total yards the away team gains:

```python
sns.scatterplot(
    data=combined_data, x="away.totalYards", y="point_differential"
).set_title("Away Team Total Yards vs Point Differential")
```

As expected, the more total yards the away team has, the average point differential gets smaller, meaning that the away team has more points than the home team.

Let's do one more analysis: look at how the difference in the total yards impacts the point differential:

```python
def total_yard_diff(row):
    return row["home.totalYards"] - row["away.totalYards"]

combined_data["total_yard_differential"] = combined_data.apply(total_yard_diff, axis=1)
sns.scatterplot(
    data=combined_data, x="total_yard_differential", y="point_differential"
).set_title("Total Yard Differential vs Point Differential")
```

This plot shows quite clearly that when the total yard differential is more positive, then that means the home team gains more total yards than the away team. This means the average point differential is higher since the home team probably has scored more points than the away team from gaining more total yards.
