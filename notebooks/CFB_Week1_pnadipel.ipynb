```python
import pandas as pd
import seaborn as sns
```

```python
game_data = pd.read_csv("../data/game_stats.csv")
games = pd.read_csv("../data/games.csv")
combined_data = games.merge(game_data, left_on="id", right_on="id", validate="one_to_one")
```

```python
def point_diff(row):
    return row["home_points"] - row["away_points"]


combined_data["point_differential"] = combined_data.apply(point_diff, axis=1)
combined_data["point_differential"]
```

```python
combined_data["home.possessionTime"]
```

```python
sns.scatterplot(data = combined_data, x="home.yardsPerRushAttempt", y="point_differential")
```

```python
sns.scatterplot(data = combined_data, x="away.yardsPerRushAttempt", y="point_differential")
```
