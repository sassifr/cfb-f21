```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
game_data = pd.read_csv("../data/game_stats.csv")
```

```python
games = pd.read_csv("../data/games.csv")
```

```python
game_data
```

```python
games
```

```python
combined_data = games.merge(game_data, left_on="id", right_on="id", validate="one_to_one")
```

```python
combined_data
```

```python
def point_diff(row):
    return row["home_points"] - row["away_points"]

combined_data["point_differential"] = combined_data.apply(point_diff, axis = 1)
```

```python
combined_data["point_differential"]
```

```python
sns.barplot(data=combined_data, x="home.tackles", 
            y="point_differential").set_title("Home Tackles v.s. Point Differential")
```

```python
sns.barplot(data=combined_data, x="away.tackles", y="point_differential").set_title("Away Tackles v.s. Point Differential")
```
