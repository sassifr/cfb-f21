```python
# MDST College Football Week 1 - Saad Monem

# Setup - Credit: skunichetty - Data Visualization Example

import pandas as pd
import seaborn as sns

game_data = pd.read_csv("../data/game_stats.csv")
games = pd.read_csv("../data/games.csv")
combined_data = games.merge(game_data, left_on="id", right_on="id", validate="one_to_one")

def point_diff(row):
    return row["home_points"] - row["away_points"]

combined_data["point_differential"] = combined_data.apply(point_diff, axis=1)

combined_data
```

```python
# Data Visualization

sns.set(rc={'figure.figsize':(10, 10)})
sns.scatterplot(
    data=combined_data, x="excitement_index", y="point_differential"
).set_title('Excitement Index vs Point Differential')
```

Explanation:

In this visualization we see an interesting pattern for excitement index. When excitement is low the point differential tends to swing to extremes, being either very high or very low. As excitement increases, the point differential converges to 0. Intuitively, this makes sense as closer games would be more exciting to watch. This tells us that we could use excitement index to predict how competitive a game was.
