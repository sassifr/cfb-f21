```python
import pandas as pd
import seaborn as sns

game_data = pd.read_csv("https://raw.githubusercontent.com/skunichetty/cfb-f21/main/data/game_stats.csv")
games = pd.read_csv("https://raw.githubusercontent.com/skunichetty/cfb-f21/main/data/games.csv")

combined_data = games.merge(game_data, left_on="id", right_on="id", validate="one_to_one")

def point_diff(row):
    return row["home_points"] - row["away_points"]


combined_data["point_differential"] = combined_data.apply(point_diff, axis=1)
combined_data["point_differential"]

sns.set(rc = {'figure.figsize':(15,8)})

sns.barplot(
    data=combined_data, x="home.firstDowns", y="point_differential"
).set_title("Home Team First Downs vs Point Differential")
```

The frequency of first downs shows how well one team can make progress towards the endzone against another team’s defense, so it would make sense that teams with more success in driving down the field tend to have higher point differentials.
